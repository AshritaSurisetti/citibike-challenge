```python
import pandas as pd
import os
```

```python
# Path to the directory containing the CSV files
data_dir = "C:\\Users\\ashri\\OneDrive\\Desktop\\Bootcamp- Assignments\\citibike-challenge\\JC data"
```

```python
for filename in os.listdir(data_dir):
    if filename.endswith(".csv"):
        # Read the first few rows of the CSV file into a DataFrame
        sample_df = pd.read_csv(os.path.join(data_dir, filename), nrows=5)
        
        print(f"Columns and data types for {filename}:")
        print(sample_df.dtypes)
        print()
```

```python

# List to store cleaned DataFrames
cleaned_dfs = []

# Iterate over each CSV file in the directory
for filename in os.listdir(data_dir):
    if filename.endswith(".csv"):
        # Read the CSV file into a DataFrame
        df = pd.read_csv(os.path.join(data_dir, filename), low_memory= False)
        
        # # Convert columns to appropriate data types
        # df['start_station_id'] = pd.to_numeric(df['start_station_id'], errors='coerce')
        # df['end_station_id'] = pd.to_numeric(df['end_station_id'], errors='coerce')
        
        # Append cleaned DataFrame to the list
        cleaned_dfs.append(df)

# Combine all cleaned DataFrames into a single DataFrame
combined_df = pd.concat(cleaned_dfs)

# Print the first few rows of the combined DataFrame
print(combined_df.head())
```

```python
print(len(combined_df))

#Find the missing values in dataframe
missing_values = combined_df.isna().sum()

# Print the count of missing values for each column
print("Missing values per column:")
print(missing_values)
```

```python
# Delete all rows with missing values
combined_df_cleaned = combined_df.dropna()

# Print the length of the cleaned DataFrame after dropping missing values
print(len(combined_df_cleaned))

# Print the count of missing values for each column after dropping
missing_values_after_drop = combined_df_cleaned.isna().sum()
print("Missing values per column after dropping:")
print(missing_values_after_drop)
```

```python
# Define the path for the output CSV file
output_csv_path = "C:\\Users\\ashri\\OneDrive\\Desktop\\Bootcamp- Assignments\\citibike-challenge\\combined_data.csv"

# Write the cleaned DataFrame to a CSV file
combined_df_cleaned.to_csv(output_csv_path, index=False)
```
